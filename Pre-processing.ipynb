```python
import pandas as pd
```

```python
df = pd.read_csv("lung_cancer_examples.csv")
df.head()
```

```python
# Rename the 'Alkhol' column to 'Alcohol'
df.rename(columns={'Alkhol': 'Alcohol'}, inplace=True)
```

```python
df['Smokes'].value_counts()
```

```python
def categorize_smokers(x):
    if x == 0:
        return 'Non-Smokers'
    elif x <= 2:
        return 'Light Smokers'
    elif x <= 10:
        return 'Mediocre Smokers'
    else:
        return 'Heavy Smokers'

def smokes(dataset):
    # Check if 'Smokes' column exists in the dataset
    if 'Smokes' in dataset.columns:
        # Apply categorize_smokers() function to 'Smokes' column and create 'Smoking_Category' column
        dataset['Smoking_Category'] = dataset['Smokes'].apply(categorize_smokers)
        return dataset
    else:
        return "Column 'Smokes' not found in the dataset."

# Apply smokes() function to categorize smoking habits and add 'Smoking_Category' column
lung_cancer_with_categories = smokes(df)
print(lung_cancer_with_categories)
```

```python
df['Alcohol'].value_counts()
```

```python
def categorize_alcohol(x):
    if x == 0:
        return 'Non-Drinkers'
    elif x <= 2:
        return 'Light Drinkers'
    elif x <= 10:
        return 'Mediocre Drinkers'
    else:
        return 'Heavy Drinkers'

def alkhol(dataset):
    # Check if 'Alcohol' column exists in the dataset
    if 'Alcohol' in dataset.columns:
        # Apply categorize_alcohol() function to 'Alcohol' column and create 'Alcohol_Category' column
        dataset['Alcohol_Category'] = dataset['Alcohol'].apply(categorize_alcohol)
        return dataset
    else:
        return "Column 'Alcohol' not found in the dataset."

# Apply alkhol() function to categorize alcohol consumption and add 'Alcohol_Category' column
lung_cancer_with_alcohol_categories = alkhol(df)
print(lung_cancer_with_alcohol_categories)
```
